2. Write a function that takes two arguments as an input and performs AND operation. The function returns the result to the main program.

```python
def andFunc(x, y):
  if( x == 1):
    if(y == 1):
      return True
  return False

first = int(input("Enter either 0 or 1: ") )
second = int(input("Enter either 0 or 1: ") )

print( andFunc(first, second) )
```

3. Write two additional functions like in part (b); each would perform OR and XOR operations and return the result to the main program.


```python
def orFunc(x, y):
  if(x == 1):
    return True
  if(y == 1):
    return True
  return False

def xorFunc(x, y):
  if(x == 1):
    if(y == 0):
      return True
  if(y == 1):
    if(x == 0):
      return True
  return False

first = int(input("Enter either 0 or 1: ") )
second = int(input("Enter either 0 or 1: ") )

print("Or operation: " ,orFunc(first, second))
print("Xor operation: " ,xorFunc(first, second) )
```

4. Create a for loop to call the three functions each for data sample each time.
5. Store the output of each function in a separate list/array.
6. Print out the results for the three functions at the end of your program

```python
andVals = []
orVals = []
xorVals = []

for x in arr:
  andVals.append(andFunc(x[0], x[1]))
  orVals.append(orFunc(x[0], x[1]))
  xorVals.append(xorFunc(x[0], x[1]))

print("Data:", arr)
print("And:", andVals)
print("Or:", orVals)
print("Xor", xorVals)
```
